```python
env = gym.make("SpaceInvaders-v0",render_mode='human')
```

```python
env.reset()
```

```python
env.render()
```

```python
action = env.action_space.sample() # From available actions sample 1 action
```

```python
# Now take that action
# If you take that action, you will get a new state (state after taking the action), reward (reward agent got), is_done/terminated (is the objective reached), truncated, info ()
new_state, reward, terminated, truncated, info = env.step(action)
```

# The new state is the frame/image for the next state; 
# Reward = 0

```python
print(new_state, reward, terminated, truncated, info)
```

```python
is_done=False
while not is_done:
    action = env.action_space.sample()
    new_state, reward, terminated, truncated, info = env.step(action)
    print(info)
    env.render() # This method updates the window based on the new_state
```

```python
from gym.wrappers.monitoring.video_recorder import VideoRecorder

env = gym.make('CartPole-v0', render_mode="rgb_array")
env.reset()
```

```python
env.render()
```

```python
is_done= False
while not is_done:
    action = env.action_space.sample()
    new_state, reward, is_done, truncated, info = env.step(action)
    print(new_state) # Array of 4 elements: [cart position, cart velocity, pole angle, pole velocity]
    print(reward) # Reward = 1.0 = constant for each action, if the episode has terminated, reward = 0
    env.render()
```

```python
n_episode = 100; total_reward_per_episode =0; 
for episode in range(n_episode):
    env.reset()
    is_done = False
    while not is_done:
        action = env.action_space.sample()
        new_state, reward, is_done, truncated, info = env.step(action)
        env.render()
        total_reward_per_episode += reward
avg_rewards = total_reward_per_episode/n_episode

print(f'Average reward per episode: {avg_rewards}')
```
